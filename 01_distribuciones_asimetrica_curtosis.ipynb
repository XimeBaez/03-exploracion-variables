### Distribucion sesgada a la derecha

```python
dist_sesgada = np.random.exponential(1, 1000)

# Uso de nuestra funcion generica
generar_distribucion(dist_sesgada, 'Distribucion Sesgada a la Derecha')
```

### Dsitribucion sesgada a la izquierda

### Distrobucion bimodal

```python
dist_bimodal = np.concatenate(
    [
        np.random.normal(-2, 1, 500), 
        np.random.normal(3, 1, 500)
    ]
)

# Uso de nuestra funcion generica
generar_distribucion(dist_bimodal, 'Distribucion Bimodal')
```

### Distribucion uniforme

```python
dist_uniforme = -np.random.uniform(-1, 1, 1000)

# Uso de nuestra funcion generica
generar_distribucion(dist_uniforme, 'Distribucion Uniforme')
```
